```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests as rq
```

```python
url = 'https://rateyourmusic.com/release/single/queen/bohemian_rhapsody___im_in_love_with_my_car/'
headers = {'User-Agent': 'bvcl'}
r = rq.get(url, headers=headers)
parsed_page = BeautifulSoup(r.text,"html.parser")
print(parsed_page)
```

```python
nomeMusica=[];
musica="asas"
for a in parsed_page.find('div',class_='album_title'):
    nomeMusica.append(str(a).strip().split("<")[0])
musica = nomeMusica[0]
print(musica)
```

```python
#pagina das reviews
#https://rateyourmusic.com/release/single/queen/bohemian_rhapsody___im_in_love_with_my_car/reviews/1/
reviewLink="https://rateyourmusic.com/release/single/queen/bohemian_rhapsody___im_in_love_with_my_car/reviews/"
linksReviewPages=["https://rateyourmusic.com/release/single/queen/bohemian_rhapsody___im_in_love_with_my_car/reviews/1"]
for b in parsed_page.find("div", {"id": "reviews_shell"}).find("div",class_="review_list").find_all('a'):
    if(b.text!=">>"):
        linksReviewPages.append(reviewLink+b.text)
print(linksReviewPages)
```

```python
url = 'https://rateyourmusic.com/release/single/queen/bohemian_rhapsody___im_in_love_with_my_car/reviews/1'
headers = {'User-Agent': 'bvcl'}
r = rq.get(url, headers=headers)
parsed_page = BeautifulSoup(r.text,"html.parser")
print(parsed_page)
```

```python
k=0
users=[]
ratings=[]
nomeMusica=[]
for a in parsed_page.find_all("span",class_='review_rating'):
    try:
        k=str(a).index("content=")
        u=str(a)[k+9]+str(a)[k+10]+str(a)[k+11]
        ratings.append(u)
    except ValueError:
        ratings.append("-")
for a in parsed_page.find_all("span",class_='review_user'):
    try:
        k=str(a).index("content=")
        users.append(str(a).split("\"")[3])
        nomeMusica.append(musica)
    except ValueError:
        users.append("-")
        
#print(users)
#print(ratings)
i=0
l=len(nomeMusica)
print(l)
while(i<l):
    if(users[i]=="-" or ratings[i]=="-"):
        users.pop(i)
        ratings.pop(i)
        nomeMusica.pop(i)
        l=l-1
    i=i+1

    
teste=pd.DataFrame({'musica':nomeMusica,
                    'usuario':users,
                    'nota':ratings,
                    })
teste
```

```python
finalDF = pd.DataFrame({'musica': [],
                        'usuario': [],
                        'nota': []
                       })
for a in linksReviewPages:
    url = str(a)
    headers = {'User-Agent': 'bvcl'}
    r = rq.get(url, headers=headers)
    parsed_page = BeautifulSoup(r.text,"html.parser")
    k=0
    users=[]
    ratings=[]
    nomeMusica=[]
    for a in parsed_page.find_all("span",class_='review_rating'):
        try:
            k=str(a).index("content=")
            u=str(a)[k+9]+str(a)[k+10]+str(a)[k+11]
            ratings.append(u)
        except ValueError:
            ratings.append("-")
    for a in parsed_page.find_all("span",class_='review_user'):
        try:
            k=str(a).index("content=")
            users.append(str(a).split("\"")[3])
            nomeMusica.append(musica)
        except ValueError:
            users.append("-")

    #print(users)
    #print(ratings)
    i=0
    l=len(nomeMusica)
    print(l)
    while(i<l):
        if(users[i]=="-" or ratings[i]=="-"):
            users.pop(i)
            ratings.pop(i)
            nomeMusica.pop(i)
            l=l-1
        i=i+1


    teste=pd.DataFrame({'musica':nomeMusica,
                        'usuario':users,
                        'nota':ratings,
                        })
    
    frames = [finalDF, teste]
    finalDF = pd.concat(frames,ignore_index=False)
finalDF
```
