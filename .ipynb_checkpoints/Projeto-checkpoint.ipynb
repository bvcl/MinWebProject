```python
userlist = []
for i in range(userArt.shape[0]):
    if userArt['userID'][i] not in userlist:
        userlist.append(userArt['userID'][i])
```

```python
evaluationDF = pd.DataFrame()
evaluationDF = evaluationDF.assign(user = userlist)
```

```python
userArt['userID'][0]
```

```python
artistlist = []
for i in range(userArt.shape[0]):
    if userArt['artistID'][i] not in artistlist:
        artistlist.append(userArt['artistID'][i])
artistlist.sort()
```

```python
evaluationDF2 = pd.DataFrame()
evaluationDF2 = evaluationDF2.assign(artist = artistlist)
```

```python
evaluationDF2.T
```

```python
resultDF = evaluationDF.append(evaluationDF2.T)
```

```python
resultDF
```

```python
state = 0
for k in range(len(userlist)):
    breakflag1 = 0
    breakflag2 = 0
    for i in range(state,userArt.shape[0]):
        if resultDF['user'][k] == userArt['userID'][i]:
            breakflag1 = 1
            breakflag2 = 0
            if userArt['artistID'][i] >= 1114:
                myrange = userArt['artistID'][i] - 1114
            else:
                myrange = 0
            for j in range(myrange, len(artistlist)):
                if resultDF[j]['artist'] == userArt['artistID'][i]:
                    resultDF[j][k] = userArt['weight'][i]
                    break
        else:
            breakflag2 = 1
        if breakflag1 and breakflag2 == 1:
            state = i
            break
    if k % 10 == 0:
        print (k)
```
