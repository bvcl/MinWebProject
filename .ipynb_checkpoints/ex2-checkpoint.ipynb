## Problema 1

Colete a tabela de classificação do site da CBF usando `requests` e `BeautifulSoup`. Você deve transformar os dados em um data frame.

```python
url = 'https://www.cbf.com.br/competicoes/brasileiro-serie-a/classificacao/2017'
headers = {'User-Agent': 'bvcl'}
r = rq.get(url, headers=headers)
parsed_page = BeautifulSoup(r.text,"html.parser")
```

```python
i=0;
aux=[];
aux2=[]
for a in parsed_page.find_all('td'):
    if(i>1):
        aux.append(a.text.strip()) 
    i=i+1
    if(i==18):
        aux2.append(dict(zip(['Classificacao','P','J','V','E','D','GP','GC','SG','VM','VV','DM','DV','CA','CV','%'],aux)))
        i=0
        aux=[]
dfTimes = pd.DataFrame(aux2,columns=['Classificacao','P','J','V','E','D','GP','GC','SG','VM','VV','DM','DV','CA','CV','%'])
dfTimes
```

## Problema 4

Plote um gráfico de dispersão para avaliar se existe alguma relação entre o investimento do clube na construção do elenco e sua pontuação (e consequente classificação) no campeonato. Caso seja possível perceber alguma relação, ajuste um modelo linear para predizer a pontuação frente ao investimento. Qual o aumento na pontuação para cada (milhão de) real investido?

```python
classificacao=[]
pontos=[]
gp=[]
gc=[]
sg=[]
percentage=[]
time_ponto_valor=[]
aux_for_df=[]
for row in dfTimes.itertuples():
    if("Atlético" in row.Classificacao):
        classificacao.append(row.Classificacao.replace(" ","").replace("-",""))
    else:
        classificacao.append(row.Classificacao.replace(" ","").split("-")[0])
    pontos.append(row.P)
    gp.append(row.GP)
    gc.append(row.GC)
    sg.append(row.SG)
    percentage.append(row._16)

i=0
for a in classificacao:
    aux_for_df.append(dict(zip(['time','pontos','percentage','valor','gp','gc','sg'],[classificacao[i],int(pontos[i]),int(percentage[i]),int(auxValores[names.index(a)]),int(gp[i]),int(gc[i]),int(sg[i])])))
    i=i+1

dfTime_pt_valor = pd.DataFrame(aux_for_df,columns=['time','pontos','percentage','valor','gp','gc','sg'])
#dfTime_pt_valor

g = dfTime_pt_valor[['valor','pontos']].plot.scatter(x='valor',y='pontos')
g.xaxis.set_major_formatter(FormatStrFormatter('€ %.2f'))
plt.show(g)
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
model = smf.ols("pontos ~ valor",data=dfTime_pt_valor)
result = model.fit()

pd.concat([result.params,result.bse,result.tvalues,result.pvalues],
          axis=1, keys=['coef','SE','t','p-value'])
```

De acordo com a tabela acima, podemos ver que, para cada euro investido, a pontuação aumenta em aproximadamente 3 * 10e-7, ou seja, para cada 10 milhões de euros, a pontuação aumenta em aproximadamente 3 pontos. Considerando o euro como 3,8 reais, temos um aumento de aproximadamente 0,8 pontos para cada 10 milhões de reais.

## Problema 5

Repita o problema 4, porém, agora, avalie a relação entre: (1) o número de gols-pró e o valor do ataque e meio-campo juntos; (2) o número de gols contra e o valor da defesa com o goleiro; (3) o saldo de gols e a diferença entre o valor do setor ofensivo (meio-campo e ataque) e o setor defensivo (defesa e goleiro). Os gráficos devem ser exibidos lado-a-lado para facilitar a análise. Ajuste modelos lineares para predizer os valores anteriores. Quem contribui mais para o número de gols a favor, ataque ou meio-campo? E para os gols contra? Defesa ou os goleiros? E, finalmente, para o saldo de gols? Defesa, ataque, meio-campo ou goleiro? Existe alguma influência nos valores investidos em conjunto em múltiplos setores?

```python
i=0
allPos=["Keeper","Centre-Back"]

for row in parsed_page_team.itertuples():
    if(str(row[2]) in allPos):
        continue
    else:
        allPos.append(row[2])
        
#print(allPos)
        
defPos=["Centre-Back",'Left-Back','Right-Back','Defence']
midPos=['Defensive Midfield','Central Midfield','Attacking Midfield','Midfield','Right Midfield','Left Midfield']
ataPos=['Left Wing', 'Right Wing', 'Centre-Forward', 'Striker', 'Secondary Striker']

gkpValues=[0]*20
defValues=[0]*20
attValues=[0]*20
midValues=[0]*20


for row in parsed_page_team.itertuples():
    if(row[2] in defPos):
        v=defValues[names.index(row[7].replace(" ","").replace("-","").replace("EC","").replace("FC","").replace("Recife",""))]
        defValues[names.index(row[7].replace(" ","").replace("-","").replace("EC","").replace("FC","").replace("Recife",""))]=v+int(row[6])
    elif(row[2] in midPos):
        v=midValues[names.index(row[7].replace(" ","").replace("-","").replace("EC","").replace("FC","").replace("Recife",""))]
        midValues[names.index(row[7].replace(" ","").replace("-","").replace("EC","").replace("FC","").replace("Recife",""))]=v+int(row[6])    
    elif(row[2] in ataPos):
        v=attValues[names.index(row[7].replace(" ","").replace("-","").replace("EC","").replace("FC","").replace("Recife",""))]
        attValues[names.index(row[7].replace(" ","").replace("-","").replace("EC","").replace("FC","").replace("Recife",""))]=v+int(row[6])  
    else:
        v=gkpValues[names.index(row[7].replace(" ","").replace("-","").replace("EC","").replace("FC","").replace("Recife",""))]
        gkpValues[names.index(row[7].replace(" ","").replace("-","").replace("EC","").replace("FC","").replace("Recife",""))]=v+int(row[6])
        
#print(keeperValues)

teste=pd.DataFrame({'time':names,
                    'gkpValues':gkpValues,
                    'defValues':defValues,
                    'midValues':midValues,
                    'attValues':attValues})
#teste
dfInfo=dfTime_pt_valor.set_index('time').join(teste.set_index('time'))
dfInfo
```

```python
f,graphics=plt.subplots(1,3, figsize=(30, 8))

gr1 = sns.regplot(x=(dfInfo["midValues"]+dfInfo["attValues"]), y=dfInfo["gp"],ax=graphics[0])
gr2 = sns.regplot(x=(dfInfo["defValues"]+dfInfo["gkpValues"]), y=dfInfo["gc"],ax=graphics[1])
gr3 = sns.regplot(x=((dfInfo["midValues"]+dfInfo["attValues"])-(dfInfo["defValues"]+dfInfo["gkpValues"])), y=dfInfo["sg"],ax=graphics[2])

gr1.set(xlabel='attAndMid')
gr1.xaxis.set_major_formatter(FormatStrFormatter('€ %.2f'))
gr2.set(xlabel='defAndGkp')
gr2.xaxis.set_major_formatter(FormatStrFormatter('€ %.2f'))
gr3.set(xlabel='difAttDef')
gr3.xaxis.set_major_formatter(FormatStrFormatter('€ %.2f'))

axes = [gr1,gr2,gr3]
plt.show(axes)
```

```python
model = smf.ols("gp ~ midValues + attValues",data=dfInfo)
result = model.fit()

pd.concat([result.params,result.bse,result.tvalues,result.pvalues],
          axis=1, keys=['coef','SE','t','p-value'])
```

Considerando a tabela acima, podemos ver que o ataque contribui mais do que o meio campo para a quantidade de gols marcados, ja que o valor do coef do ataque é aproximadamente 3,75 vezes o do meio campo. Isso é, para uma mesma quantia investida em ambos os setores, o ataque dará um retorno 3,75 vezes maior em gols. Por exemplo, para uma quantia de 10 milhões de euros o ataque iria contribuir com 4,5 gols e o meio com 1,2 gols aproximadamente.

```python
model = smf.ols("gc ~ gkpValues + defValues",data=dfInfo)
result = model.fit()

pd.concat([result.params,result.bse,result.tvalues,result.pvalues],
          axis=1, keys=['coef','SE','t','p-value'])
```

Considerando a tabela acima, podemos ver que o goleiro contribui mais do que a defesa para a diminuir a quantidade de gols sofridos. Nesse caso, a diferença é de aproximadamente 10 vezes. Isso é, para uma mesma quantia investida em ambos os setores, o goleiro representa uma redução no numero de gols sofridos 10 vezes maior do que a defesa. Por exemplo, para uma quantia de 10 milhões de euros o goleiro iria contribuir com uma redução de 23 gols e a defesa com uma redução de 2,1 gols aproximadamente.

```python
model = smf.ols("sg ~ attValues + midValues + gkpValues + defValues",data=dfInfo)
result = model.fit()

pd.concat([result.params,result.bse,result.tvalues,result.pvalues],
          axis=1, keys=['coef','SE','t','p-value'])
```

Considerando a tabela acima, podemos ver que o goleiro é quem mais contribui para o aumento do saldo de gols. Isso já era esperado pelos resultados obtidos anteriormente, nos quais vimos que 10 milhões de euros resultariam em -23 gols sofridos. Nessa tabela podemos ver que os goleiros aumentariam em 31 o saldo de gols enquanto outros setores chegariam a um aumento de no máximo 6,5. Um fato curioso é que o coef do meio campo é negativo, ou seja investir dinheiro no meio campo resulta em uma diminuição no saldo de gols.

## Problema 6

Ajuste um modelo linear usando o valor investido em cada setor para predizer o número de pontos conquistados pela equipe. O que se pode concluir em relação ao custo benefício desses investimentos?

```python
dfPt_By_Sector = dfInfo.drop('valor', 1).drop('gp',1).drop('gc',1).drop('sg',1)
dfPt_By_Sector.corr()
```

```python
model = smf.ols("pontos ~ attValues + midValues + gkpValues + defValues",data=dfPt_By_Sector)
result = model.fit()

pd.concat([result.params,result.bse,result.tvalues,result.pvalues],
          axis=1, keys=['coef','SE','t','p-value'])
```

De acordo com a tabela, podemos ver que a pontuação base é de 28 pontos e um investimento de 10 milhões de euros resulta em:

    -5 pontos a mais para o ataque
    
    -1.35 pontos a mais para a defesa
    
    -21 pontos a mais para os goleiros
    
    -0,7 pontos a menos para o meio campo
    
  Assim, podemos que o setor com melhor custo x benefício é o de goleiros.
  

## Problema 7

Ajuste um modelo para predizer, dada as características da equipe, se ela classificará para a Copa Libertadores, ou para a Copa Sul-Americana, será rebaixada para a Série B, ou apenas se manterá na Série A.

```python
#Tabela que mostra relacao entre percentual de aproveitamento e os valores investidos por setor
#O percentual será usado para se ter uma ideia de quanto de aproveitamento será necessário para obter uma classificação
normalPercent=[]
namesOrdered=[]
vitorias=[]
empates=[]
derrotas=[]
dfInfo.drop('percentage',1)
for r in dfTimes.itertuples():
    vitorias.append(int(r.V))
    empates.append(int(r.E))
    derrotas.append(int(r.D))
    
for r in dfInfo.itertuples():
    normalPercent.append(r.percentage/100)

for r in dfInfo.itertuples():
    namesOrdered.append(r.Index)
    
t=pd.DataFrame({'time':namesOrdered,
                'normalPercent':normalPercent,
                'vitorias':vitorias,
                'empates':empates,
                'derrotas':derrotas})

dfInfo=dfTime_pt_valor.set_index('time').join(t.set_index('time'))

dfInfoShuffled = dfInfo.sample(frac=1)
dfInfoShuffled
```

```python
part1=dfInfoShuffled[0:14]
part2=dfInfoShuffled[14:19]
#part1
model = smf.logit("normalPercent ~ derrotas + empates + vitorias",part1,).fit()
model.summary()
```

```python
f = model.predict(part2)
for r in f:
    if(r>0.51):
        print("Liberta")
    elif(r>=0.42):
        print("Sul-Americana")
    elif(r>=0.36):
        print("Fica na A")
    else:
        print("Vai pra B")
```
